```python
                                                        # numpy project 1  # Temperature Sheet
import numpy as n
#1. Represent the data in the given sheet into an appropriate NumPy array so that you can perform the following actions on it.
#  creating array for november month
nov=n.array([[[[9],[26]],[[7],[27]],[[13],[26]],[[11],[25]],[[14],[25]],[[10],[32]],[[8],[21]]],[[[10],[24]],[[14],[26]],[[7],[28]],[[-13],[32]],[[6],[21]],[[11],[24]],[[13],[30]]],[[[15],[27]],[[6],[28]],[[10],[25]],[[15],[29]],[[14],[27]],[[7],[31]],[[14],[31]]],[[[14],[31]],[[11],[22]],[[7],[28]],[[14],[30]],[[6],[28]],[[11],[32]],[[12],[22]]]])
#creating array for december month
dec=n.array([[[[9],[26]],[[13],[27]],[[7],[27]],[[6],[31]],[[8],[32]],[[15],[26]],[[8],[23]]],[[[9],[31]],[[10],[25]],[[12],[20]],[[14],[29]],[[10],[30]],[[9],[24]],[[10],[20]]],[[[10],[26]],[[9],[32]],[[6],[24]],[[7],[23]],[[10],[22]],[[10],[30]],[[10],[32]]],[[[7],[26]],[[14],[23]],[[15],[25]],[[12],[26]],[[12],[31]],[[8],[27]],[[14],[32]]]])
#creating array for jan month
jan=n.array([[[[4],[19]],[[8],[19]],[[3],[21]],[[3],[20]],[[5],[19]],[[8],[5]],[[20],[5]]],[[[2],[21]],[[3],[20]],[[2],[18]],[[4],[19]],[[2],[19]],[[4],[6]],[[22],[6]]],[[[8],[22]],[[4],[22]],[[8],[18]],[[5],[21]],[[2],[20]],[[8],[3]],[[21],[4]]],[[[4],[21]],[[5],[21]],[[2],[19]],[[4],[19]],[[8],[22]],[[8],[7]],[[19],[2]]]])
# creating array for feb month
feb=n.array([[[[9],[18]],[[10],[19]],[[-10],[19]],[[8],[20]],[[7],[20]],[[10],[71]],[[20],[12]]],[[[7],[20]],[[7],[18]],[[7],[19]],[[7],[22]],[[9],[19]],[[10],[11]],[[19],[8]]],[[[9],[18]],[[12],[19]],[[8],[20]],[[11],[20]],[[11],[20]],[[10],[9]],[[18],[19]]],[[[11],[20]],[[7],[22]],[[12],[19]],[[7],[21]],[[10],[20]],[[10],[10]],[[19],[11]]]])
# creating array of complete sheet
temp=n.array([nov,dec,jan,feb])

#2. Write the dimensions and shape of the NumPy array that you havecreated.
temp.ndim   #dimension = 5
temp.shape  #shape =(4,4,7,2,1)

#3. Print the daily temperatures for the first week of each month
for month_idx, month_data in enumerate(temp):
    print(f"Month {month_idx + 1}:")
    first_week_temperatures = month_data[:, :7, :, 0]
    print(first_week_temperatures)
    print("----------------------")
    
# Task 4: Print the temperatures for Tuesday of each month.
tuesday_temperatures = temp[:, :, 1, 1, 0]
print("Temperatures for Tuesdays:")
print(tuesday_temperatures)
print("----------------------")

# Task 5: Print only the maximum temperature for all the weekdays of Dec and Feb.
dec_max_temperatures = temp[1, :, :, :, 0].max(axis=1)
feb_max_temperatures = temp[3, :, :, :, 0].max(axis=1)
print("Max temperatures for weekdays in December:")
print(dec_max_temperatures)
print("Max temperatures for weekdays in February:")
print(feb_max_temperatures)
print("----------------------")

# Task 6: Print all the days along with the week number in November when the minimum temperature was less than 8 degrees.
nov_min_below_8 = n.where(temp[0, :, :, :, 0] < 8)
print("Days and weeks in November with min temp below 8 degrees:")
print(nov_min_below_8)
print("----------------------")

# Task 7: Print all the weeks in Dec and Jan where the maximum temperature has crossed a threshold of 20 degrees.
dec_jan_max_above_20 = n.where((temp[1, :, :, :, 0] > 20) | (temp[2, :, :, :, 0] > 20))
print("Weeks in December and January with max temp above 20 degrees:")
print(dec_jan_max_above_20)
print("----------------------")

# Task 8: Check if there are any absurd values present in the dataset
absurd_values = temp[(temp < -10) | (temp > 50)]
print("Absurd values in the dataset:")
print(absurd_values)
print("----------------------")

# Task 9: What strategy would you use to handle such data points?
# Handling absurd values can involve several strategies, such as removing them, replacing with NaN, interpolation, etc.

# Task 10: Find and print the indexes of all the outlier(unusual) values present in the above dataset.
outlier_indices = n.where((temp < -5) | (temp > 40))
print("Indices of outlier values:")
print(outlier_indices)
print("----------------------")

# Task 11: Replace the outliers with an appropriate value, e.g., the average of non-outliers.
# Calculate the mean of non-outlier values
non_outliers = temp[(temp > -5) & (temp < 40)]
mean_non_outliers = n.mean(non_outliers)

# Replace outlier values with the calculated mean of non-outliers
temp[outlier_indices] = mean_non_outliers

# Display the updated array with outliers replaced
print("Updated array with outliers replaced:")
print(temp)
print("----------------------")

# Task 12: Find the average max temperature for the winter months in Jaipur.
winter_months = temp[1:3]  # Dec and Jan
max_temps = winter_months[:, :, :, 1, 0]
avg_max_temp_winter = n.nanmean(max_temps)
print("Average max temperature for the winter months in Jaipur:", avg_max_temp_winter)
print("----------------------")

# Task 13: Find the weekly min avg temp for the month of Dec in Jaipur.
dec_avg_temp_weekly = n.nanmean(temp[1, :, :, :, 0], axis=1)
print("Weekly average temperature for the month of December in Jaipur:")
print(dec_avg_temp_weekly)
print("----------------------")

# Task 14: Find the overall avg temp for the months Dec and Jan.
dec_jan_avg_temp = n.nanmean(temp[1:3, :, :, :, 0])
print("Overall average temperature for the months of December and January:", dec_jan_avg_temp)
print("----------------------")

# Task 15: Find the least temp experienced by the city in the month of Dec and Jan.
dec_least_temp = n.nanmin(temp[1, :, :, :, 0])
jan_least_temp = n.nanmin(temp[2, :, :, :, 0])
print("Least temperature in December:", dec_least_temp)
print("Least temperature in January:", jan_least_temp)
print("----------------------")

# Task 16: Find the max temp in the month of Feb and return its date (Day/Week/Month).
feb_max_temp = n.nanmax(temp[3, :, :, :, 0])
feb_max_temp_indices = n.where(temp[3, :, :, :, 0] == feb_max_temp)
print("Max temperature in February:", feb_max_temp)
print("Indices of max temperature in February:", feb_max_temp_indices)
print("----------------------")

# Task 17: Find the days in the month of Nov where the max temp of the day dropped below the avg temp of the month.
nov_avg_temp = n.nanmean(temp[0, :, :, :, 0])
nov_days_below_avg = n.where(temp[0, :, :, :, 0] < nov_avg_temp)
print("Days in November where max temp dropped below average:", nov_days_below_avg)
print("----------------------")

# Task 18: Convert the above dataset into an array where the weeks of the same month are in the same row.
reshaped_temp = temp.reshape(4, -1, 2, 1)  # Reshape to have weeks of the same month in the same row
print("Reshaped array with weeks of the same month in the same row:")
print(reshaped_temp)
print("----------------------")

# Task 19: Convert the temperature data from Celsius to Fahrenheit
temp_fahrenheit = temp * 9/5 + 32
print("Temperature data in Fahrenheit:")
print(temp_fahrenheit)
print("----------------------")

# Task 20: Sort the above data in descending order on the basis of weekly average for the month of Dec.
dec_avg_weekly_temp = n.nanmean(temp[1, :, :, :, 0], axis=1)
dec_sorted_indices = n.argsort(dec_avg_weekly_temp)[::-1]  # Sort in descending order
sorted_dec_temp = temp[1, dec_sorted_indices]
print("Sorted temperature data for December based on weekly average:")
print(sorted_dec_temp)
print("----------------------")

# Task 21: Sort the temp of the first three days of each month in descending order
# on the basis of overall average for the whole winter.

# Extract temperature data for the first three days of each month
first_three_days = temp[:, :, :3, :, 0]

# Calculate the overall average for the whole winter
winter_avg = n.nanmean(first_three_days)

# Calculate the average temperature for the first three days of each month
avg_temp_first_three_days = n.nanmean(first_three_days, axis=(2, 3))

# Create an array of indices that will allow sorting based on winter_avg
sort_indices = n.argsort(avg_temp_first_three_days[:, 0])

# Sort the first_three_days array based on the sorting indices
sorted_first_three_days = first_three_days[sort_indices]

print("Sorted temperature data for the first three days of each month:")
print(sorted_first_three_days)
print("----------------------")

# Task 22: Create an array that stores the difference between the min and max temp for each day in all the winter months.
min_temp = n.nanmin(temp, axis=4)
max_temp = n.nanmax(temp, axis=4)
temp_diff = max_temp - min_temp

print("Difference between min and max temperature for each day in all winter months:")
print(temp_diff)
print("----------------------")

# Task 23: Find and store the difference between the max temp of two consecutive days for each month of winter season.
max_temp_diff_monthly = n.diff(temp, axis=3)

print("Difference between max temperature of two consecutive days for each month:")
print(max_temp_diff_monthly)
print("----------------------")

# Task 24: Find and store the difference between the minimum temp of two consecutive days for each month of the winter season.
min_temp_diff_monthly = n.diff(temp, axis=3)

print("Difference between min temperature of two consecutive days for each month:")
print(min_temp_diff_monthly)
print("----------------------")

# Task 25: Create an array by combining the data present in arrays created in q.23 and q.24
combined_diff = n.concatenate((max_temp_diff_monthly, min_temp_diff_monthly), axis=4)

print("Combined difference between min and max temperature of each day for all winter months:")
print(combined_diff)
print("----------------------")












```
